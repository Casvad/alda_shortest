```python
with open("out.txt", "r") as file:
    data = file.read()
    data = data.replace("[", "").replace("]", "").split("\n")
    data.pop()
data = [list(map(int, line.strip("").split(", "))) for line in data]
```

```python
data
```

```python
import matplotlib.pyplot as plt

binary = [item[1] for item in data]
linear = [item[2] for item in data]
ternary = [item[3] for item in data]
time_values = [item[0] for item in data]

plt.figure(figsize=(10, 6))
plt.plot(time_values, binary, label="Binary Search", marker="o")
plt.plot(time_values, linear, label="Linear Search", marker="o")
plt.plot(time_values, ternary, label="Ternary Search", marker="o")


plt.xlabel("Dataset size")
plt.ylabel("Time (s)")
plt.title("Sorting comparison")
plt.legend()
plt.grid(True)
plt.show()
```
